```python
# Works well with pandas dataframe objects to statistically plot beautiful styles.
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
# run a single plot of the total bills of the tips
# this is a histogram and kde(kernel density extimation- the line)
# sns.distplot(tips['total_bill'])

# for just the histogram with the inner line use the code below
# sns.displot(tips['total_bill'])

# This is the histogram without the lines. it is a distribution of the total bills against the count
# bins plots the single instances of the total bills for every price points. The higher the bin, the more detailed and clumsy
# the histogram looks

sns.histplot(tips['total_bill'], bins=30)
```

```python
# sns joint plot allows the comparison of two dataset
sns.jointplot(x='total_bill', y='tip', data=tips, kind="hex")
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind="reg")
# shows the linear regression
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind="kde")
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# pair wise relationship can be plotted across a numerical dataframe. It is a joint plot of all numerical dataframe
# it is a nice way to visualize data. it accepts hue argument that are categorical eg sex
# pallette allows to colour with some specific colors
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# It is a simple concept that draws a dash mark for every point on the univarant(one single variable). kde is the sum of the rugplot
sns.rugplot(tips['total_bill'])
```

```python
# CATEGORICAL PLOTS
import seaborn as sns

# to see the visualizations inside notebook
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(30)
```

```python
# aggregate the categorical data based on mean by default
# Average total bills for males is more than females
# estimator is an attribute that hold statistcal functions to run on the plot
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# estimator is explicitly counting no of occurences. its like pandas count method. Y axis is chosen so set the x
sns.countplot(x='sex', data=tips)
```

```python
# its used to show a distributon of quaitative data that compares variables distribution.
# This shows the median, mean, interquartile ranges and outlier(extreme data points) of variable distribution.
sns.boxplot(x='day', y='total_bill', data=tips, hue="smoker")
```

```python
# Violin plots: It is harder to read but gives more information.
# x is the category, y is numerical column
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips,hue='sex')
```

```python
# A combinatin of a violin and strip plot. It shows the distribution on each stack horizonatally
# They dont sacle well for large data sets.
# This shows a combination of violin and swarm plot.
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# category plot : have an attribute such as bar,strip,violin
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
# MATRIX PLOTS for heatmaps
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flight = sns.load_dataset('flights')
tips.head()
```

```python
flight.head()
```

```python
tc = tips.corr(numeric_only=True)
sns.heatmap(tc,annot = True, cmap='coolwarm')
```

```python
fp = flight.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=1)
```

```python
# cluster map: this cluster rows and cols based on their similarity
import scipy
import seaborn as sns
%matplotlib inline

flight = sns.load_dataset('flights')
fp = flight.pivot_table(index='month', columns='year', values='passengers')
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

```python
# GRIDs PLOTS: automate some plots based on features in our dataset
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
# measurement of different species of flowers
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# for more control of the actual plot
import matplotlib.pyplot as plt
grid = sns.PairGrid(iris)
grid.map(plt.scatter)
```

```python
grid = sns.PairGrid(iris)
grid.map_diag(sns.histplot)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```
