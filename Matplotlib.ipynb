```python
# It is a plotting library for python for data visualization and same feel for matlab graphical plotting.
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
# Functional method
plt.plot(x,y, 'r-')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('x and y correlation')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Matploylib object oriented API method means instantiate figure objects and call methods on those instances
# OOP: figure object is like an imaginary blank canvass
fig = plt.figure()
# left, bottom, width and height
axes = fig.add_axes([0.3, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('XY Title')
```

```python
fig = plt.figure()
# axe2-width = ax1-width - (ax2-left - ax1-left)
# axe2-height = ax1-height - (ax2-bottom - ax1-bottom)
# OR
# ax2-left = ax1-W + ax1-L - ax2W
# ax2-bottom = ax1-H + ax1-B - ax2H
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
# create subplots using oop
fig, axes = plt.subplots(nrows=1, ncols=2)

# this gives space between overlapping plots
# plt.tight_layout()

# since axes is an array of subplot axe, we can iterate the items
# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x,y,'r')
axes[0].set_title('First Plot')

axes[1].plot(y,x,'b')
axes[1].set_title('First Plot')

plt.tight_layout()

# axes.plot(x,y)
```

```python
# Figure size and DPI
# (5,2 )means 5inches horizontal  by 2inches vertical. DPI = dot per inch or pixel
# fig = plt.figure(figsize=(5,2), dpi=100)
fig = plt.figure(figsize=(8,3))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# create multiple plots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5,2))
axes[0].plot(x,y)
axes[1].plot(y,x,'b')
plt.tight_layout()
```

```python
# How to save a figure using svg, pdf, png,jpeg
fig.savefig('my_picture.png', dpi=200)
```

```python
# create a single graph plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_title('X-Y Graph')
axes.set_xlabel('X-side')
axes.set_ylabel('Y-side')
```

```python
# legend in graphs
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2 , label='x-squared')
axes.plot(x,x**3, label='x-cubed')

# labels refrences the names of the plots in the graph and loc=0 for best location for the legend
axes.legend(loc=0)

# loc(0.1, 0.1) is a location like what we did with graphs
# axes.legend(loc=(0.1, 0.1))
```

```python
# PLOT APPEARANCE
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# use hex code colors, alpha=opacity of line, linewidth=thickness of line, linestyle = dot,steps or single line
# marker marks each points of x on y(o, 1,+,*)
ax.plot(x,y, color="#FF8C00", linewidth=2, alpha=0.5, linestyle='--', marker = "+", markersize="15", markerfacecolor="blue", markeredgewidth=2, markeredgecolor="green") 
```

```python
# Control over axis appaerance in plot range
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple",lw=2, ls="--")

# to show the plot btw 0 and 1 on the x axis
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
# Matplotlib assignment
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x*20,y*10)
axes.set_xlabel('X-side')
axes.set_ylabel('Y-side')
axes.set_title('XY Graph')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x*20,y*400)
ax2.plot(x,y)

ax2.set_xlim([20,22])
ax2.set_ylim([30,50])

ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```
